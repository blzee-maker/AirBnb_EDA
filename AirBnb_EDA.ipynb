```python
#importing required library's
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#mounting the data.
df=pd.read_csv('automobile.csv')
df
```

```python
df.shape
```

```python
df.columns
```

```python
#replacing the unwanted or entity from the data set

df.replace('?',np.nan,inplace=True)#replace the ? with NAN
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#Finding the null values.

df.isna().sum()
```

```python
#droping rows with missing values in the price column
df.dropna(subset=['price'],axis=0,inplace=True)
```

```python
#original Shape:(205,26)
df.shape#This indicates that 4 rows have been droped with null values in the price column
```

```python
#replacing the missing values from the normalized-losses column with its mean
df['normalized-losses'].fillna(df['normalized-losses'].astype(float).mean(),inplace=True)
```

```python
df[['normalized-losses','bore','stroke','horsepower','peak-rpm','price']]=df[['normalized-losses','bore','stroke','horsepower','peak-rpm','price']].astype(float)
```

```python
df.info()
```

```python
#checking for duplicate rows.
df.drop_duplicates(inplace=True)
df.shape #having same values indicate that there are no duplicate values in the data set.
```

```python
#Creating a box plot for price Feature
sns.boxplot(x="price",data=df)

#Displaying the Plot
plt.show()
```

```python
#Dealing with the outliers

df=df[(df['price'] >= df['price'].quantile(0.05)) & (df['price'] <= df['price'].quantile(0.95))]
```

```python
df.shape #this Indicates that 20 outliers were removed
```

```python
sns.boxplot(x="price",data=df)

#Displaying the Plot
plt.show()

#we can see a significant decrese in the outliers.
```
